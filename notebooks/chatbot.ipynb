```python
import os
import nltk
import ssl
import streamlit as st 
import random
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model  import LogisticRegression

ssl._create_default_https_context = ssl._create_unverified_context
nltk.data.path.append(os.path.abspath("nltk_data"))
nltk.download('punkt')
```

```python
intents = [
    {
        "tag": "greeting",
        "patterns": ["Hi", "Hello", "Hey","How are you","What's up"],
        "responses": ["Hi there", "Hello", "Hey","I'm fine, thank you","Nothing much"]
    },
    {
        "tag": "goodbye",
        "patterns": ["Bye", "See you later", "Goodbye","Take care"],
        "responses": ["GoodBye", "See you later", "Take care"]
    },
    {
        "tag": "thanks",
        "patterns": ["Thank you", "thanks", "Thanks a lot","I appreciate it"],
        "responses": ["You're welcome", "No problem", "Glad I could help"]
    },
    {
        "tag": "about",
        "patterns": ["What can you do", "who are you", "what are you","What is your purpose","What's up"],
        "responses": ["I am a chatbot", "My purpose is to assist you", "I can answer questions and provide assistance"]
    },
    {
        "tag": "about",
        "patterns": ["What can you do", "who are you", "what are you","What is your purpose","What's up"],
        "responses": ["I am a chatbot", "My purpose is to assist you", "I can answer questions and provide assistance"]
    },{
        "tag": "help",
        "patterns": ["Help", "I need help", "Can you help me","What should i do"],
        "responses": ["Sure, What do you need help with?", "I'm here to help. What's the problem?", "How can I assist","What do you need help for?"]
    },{
    "tag": "age",
    "patterns": ["What's my age?", "Can you tell me my age?", "How old am I?", "Calculate my age", "Age calculator"],
    "responses": ["I can help with that! Please provide your birth year.", "Tell me your birth date, and I'll calculate your age.", "Sure, let me know when you were born!", "I need your birth year to calculate your age."]
    },
    {
    "tag": "weather",
    "patterns": ["What's the weather?", "How's the weather today?", "Tell me the forecast", "Is it sunny today?", "Is it raining?"],
    "responses": ["The current weather depends on your location. Where are you?", "Let me know your location, and I'll provide the weather.", "I can help with the forecast. Where are you located?", "Sure, I'll check the weather for you!"]
    },
   {
    "tag": "budget",
    "patterns": ["How do I manage my budget?", "I need budget tips", "Help with budgeting", "How do I create a budget?", "Budget planner"],
    "responses": ["Budgeting is important! Start by tracking your income and expenses.", "I can guide you to create a budget. What’s your monthly income?", "Budgeting starts with setting priorities. What are your financial goals?", "Need tips on managing a budget? Let’s discuss your spending habits."]
    },
   {
    "tag": "credit_score",
    "patterns": ["What is a credit score?", "How do I check my credit score?", "Why is my credit score important?", "How to improve credit score?", "Credit score tips"],
    "responses": ["A credit score is a number that reflects your creditworthiness.", "To improve your credit score, make payments on time and reduce your debts.", "You can check your credit score through financial apps or credit bureaus.", "Understanding your credit score is essential for financial planning. Need more help?"]
    }

]
```

```python
vectorizer = TfidfVectorizer()
clf = LogisticRegression(random_state=0, max_iter=100000)

tags = []
patterns = []
for intent in intents:
    for pattern in intent['patterns']:
        tags.append(intent['tag'])
        patterns.append(pattern)
        
x = vectorizer.fit_transform(patterns)
y = tags
clf.fit(x,y)
```

```python
from scipy.sparse import csr_matrix

sparse_matrix = csr_matrix(x)
print(sparse_matrix)
```

```python
def chatbot(input_text):
    input_text = vectorizer.transform([input_text])
    tag = clf.predict(input_text)[0]
    for intent in intents:
        if intent['tag'] == tag:
            response = random.choice(intent['responses'])
            return response
    
```

```python
chat_1 = chatbot("Hello")
print(chat_1)
```

```python
counter = 0

def main():
    global counter
    st.title("Chatbot")
    st.write("Welcome to the chatbot. Please type a message and press Enter to start the conversation.")
    
    counter += 1
    user_input = st.text_input("You:" , key=f"user_input_{counter}")
    
    if user_input:
        response = chatbot(user_input)
        st.text_area("Chatbot:", value=response, height=100, max_chars=None, key=f"chatbot_response_{counter}")
        
        if response.lower() in ['goodbye','bye']:
            st.write("Thank you for chatting with me.Have a hreat day!")
            st.stop()
            
if __name__ == '__main__':
    main()
```
